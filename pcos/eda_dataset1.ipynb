## Studying the features

For every Yes/No questions:  
- Yes = 1
- No = 0

```python
# Binary variables selection
binary_variables = [
    "Pregnant(Y/N)",
    "Weight gain(Y/N)",
    "hair growth(Y/N)",
    "Skin darkening (Y/N)",
    "Hair loss(Y/N)",
    "Pimples(Y/N)",
    "Fast food (Y/N)",
    "Reg.Exercise(Y/N)",
]
data[binary_variables] = data[binary_variables].astype(int)
binary_data = data[binary_variables]

# How many 0 and 1 per variables
count_data = binary_data.apply(lambda x: x.value_counts()).T

# The plot
ax = count_data.plot(kind="bar", figsize=(16, 5))

# Labels
plt.xlabel("Binary variables")
plt.ylabel("Count")
plt.title("Answers distribution for each binary variable")

# PManaging the label of the binary variables
ax.set_xticklabels(binary_variables, rotation=45, ha="right")

# Afficher le graphique
plt.legend(title="Label", loc="upper left", labels=["0 = No", "1 = Yes"])
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(30, 5))
sns.countplot(data, x="Cycle(R/I)", ax=axs[0])
sns.countplot(data, x=" Age (yrs)", ax=axs[1])
sns.countplot(data, x="Blood Group", ax=axs[2])
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(30, 5))
sns.histplot(data["Weight (Kg)"], ax=axs[0])
sns.histplot(data["Height(Cm) "], ax=axs[1])
sns.histplot(data["BMI"], ax=axs[2])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 5))
sns.boxplot(
    data=data[["Avg. F size (L) (mm)", "Avg. F size (R) (mm)", "Endometrium (mm)"]]
)
plt.title("The average size of follicles and the endometrium thickness in display")
plt.xlabel("Type of measurement")
plt.ylabel("Measure in mm")
plt.xticks(
    [0, 1, 2],
    [
        "Avg size of follicles in left Ovary",
        "Avg size of follicules in right Ovary",
        "Endometrium thickness",
    ],
)
plt.show()
```

```python
f, axs = plt.subplots(1, 2, figsize=(12, 4), gridspec_kw=dict(width_ratios=[4, 3]))
# Scatter plot (number of follicles)
sns.scatterplot(
    data=data, x="Follicle No. (L)", y="Follicle No. (R)", hue="PCOS (Y/N)", ax=axs[0]
)
# Histogramme (ages)
sns.histplot(
    data=data,
    x="PCOS (Y/N)",
    hue="PCOS (Y/N)",
    shrink=0.8,
    alpha=0.8,
    legend=False,
    ax=axs[1],
)

f.tight_layout()
```

## Equilibrium of target classes

```python
y_distribution = data["PCOS (Y/N)"].value_counts()
plt.figure(figsize=(6, 4))
y_distribution.plot(kind="bar")
plt.title("Distribution of Y classes")
plt.xlabel("PCOS (Y/N)")
plt.ylabel("Number of patients")
plt.xticks(rotation=0)
plt.show()

# Afficher les proportions des classes
print(y_distribution)
```

The classes of the target variable are unbalanced.

## Correlation between target and features

```python
data.describe()
```

```python
data_extract = data.iloc[:, 2:]
plt.figure(figsize=(16, 6))
# Isolating the values of the upper triangle:
    # changing it to 1 and changing the lower ones to 0
mask = np.triu(np.ones_like(data_extract.corr(), dtype=np.bool))
heatmap = sns.heatmap(data_extract.corr(), mask=mask, vmin=-1, vmax=1, cmap='PiYG')
heatmap.set_title('Features correlation to PCOS diagnosis', fontdict={'fontsize':18}, pad=16);
```

```python
#correlation of all features to the PCOS diagnosis

plt.figure(figsize=(7, 10))
heatmap = sns.heatmap(data_extract.corr()[['PCOS (Y/N)']].sort_values(by='PCOS (Y/N)', ascending=False), vmin=-1, vmax=1, annot=True, cmap='PiYG')
heatmap.set_title('Features\'s correlation to the PCOS diagnosis', fontdict={'fontsize':18}, pad=16);
```
